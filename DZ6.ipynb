## Задача 1 
Продакт на главной mail.ru решил протестировать в рекомендательной ленте контента вместо карточек со статьями видеоплеер с короткими видео. Нынешний таймспент на юзера в день в среднем равен 25 минут, а стандартная ошибка (SD) равна 156. Мы предполагаем, что в новой версии таймспент на юзера в день изменится на 10%. Средний трафик 20000 человек в день. Посчитайте сколько дней необходимо держать эксперимент при alpha = 5% и beta = 20% .



### Определение размера выборки для сравнения двух независимых средних
- Расчитаем ожидаемый таймспент на юзера в день: $25+25*10\%=27.5$  
- Используем [Калькулятор](https://statulator.com/SampleSize/ss2M.html)  
Предполагая, что объединенное стандартное отклонение составляет 156 единиц, для исследования потребуется размер выборки: 
61124 для каждой группы (т. е. общий размер выборки 122 248 при условии одинакового размера групп) для достижения мощности 80 % и уровня значимости 5 % (двусторонний) для выявления истинной разницы в средних значениях между тестом и тестом. контрольная группа 2,5 (т.е. 27,5 - 25) единиц

### Определение продолжительности зксперемента
$122248/20000 \approx 6.1124 = \color{green}7\space дней$

## Задача 2 
Наша продуктовая команда в ecommerce магазине планирует запустить тест, направленный на ускорение загрузки сайта. Одна из основных метрик bounce rate в GA = 40%. Мы предполагаем, что при оптимизации сайта она изменится минимум на 20%.Средний трафик 4000 человек в день. Посчитайте сколько нам нужно дней держать эксперимент при alpha = 5% и beta = 20%

### Определение размера выборки 
- Используем [Калькулятор](https://www.evanmiller.org/ab-testing/sample-size.html#!40;80;5;20;1) определяем, что для исследования потребуется размер выборки $592$ на вариант.  
Итого $1184$  


### Определение продолжительности зксперемента
$1184/4000 \approx 0.296 = \color{green}1\space день$
